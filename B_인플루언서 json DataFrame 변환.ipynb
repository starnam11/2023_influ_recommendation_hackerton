```python
# 사용할 모듈 import

import json, os
import pandas as pd
```

```python
# 각 인플루언서 별 data.json 파일 접근

'''
1. '인플루언서정보_샘플' 폴더에 있는 100명의 폴더 이름 추출 -> 리스트 형태로 저장
2. 개별 인플루언서 폴더에서 'data.json' 불러오기
3. json 파일을 DataFrame으로 변환 후, csv 파일로 내보내기
'''

path = '/content/drive/MyDrive/2023 KUIAI 해커톤/Data/인플루언서정보_샘플/'

# 인플루언서 인스타그램 id로 된 폴더 이름 추출 -> 리스트 형태로 저장
inf_id = os.listdir(path)

# 인플루언서 폴더 경로
for i in range(len(inf_id)):
  path_inf = path + inf_id[i]

  # data.json -> DataFrame으로 변환
  with open(path_inf + '/data.json') as f:
    js = json.loads(f.read())
  df = pd.DataFrame(js)

  # DataFrame에서 필요없는 column('timestamp') 제거
  df = df.drop(columns = 'timestamp')

  # DataFrame -> csv파일로 저장 (개별 인플루언서 폴더에)
  df.to_csv(path_inf + '/json_data.csv')
```
