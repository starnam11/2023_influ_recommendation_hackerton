무신사 이미지로 학습된 Style, Tone, Season Classifier들을 이용, 인플루언서들의 피드 이미지를 input으로 대입하여 각 피드 이미지의 Style, Tone, Season 태그를 추출한 후, 정보를 Dataframe으로 저장하는 코드임.

```python
# 필요한 모듈 import

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing import image
from tensorflow.keras.layers import GlobalAveragePooling2D
import pandas as pd
import pathlib as Path
import os
from tensorflow.keras.applications.inception_resnet_v2 import InceptionResNetV2, preprocess_input
```

# 남자 Style

```python
# 모델 로드
model_man_style = tf.keras.models.load_model('/content/drive/MyDrive/man_style_classificator_brand_new.h5')
```

```python
# 빈 DataFrame 생성
df_influ_style = pd.DataFrame(columns = ['influ_id', 'influ_pic_id','style'])
df_influ_style
```

```python
root_file_path = os.listdir('/content/drive/MyDrive/2023 KUIAI 해커톤/influ_man_half')
```

```python
df_influ_style['influ_pic_id'] = root_file_path
```

```python
for i in range(len(df_influ_style)):
  dir_path = '/content/drive/MyDrive/2023 KUIAI 해커톤/influ_man_half/'+df_influ_style['influ_pic_id'][i]
  test_image = image.load_img(dir_path, target_size=(224,224))
  test_image = image.img_to_array(test_image)
  
  test_image = test_image.reshape((1, test_image.shape[0], test_image.shape[1], test_image.shape[2]))
  test_image = preprocess_input(test_image)
  prediction = model_man_style.predict(test_image)

  df = pd.DataFrame({'pred':prediction[0]})
  df = df.sort_values(by='pred', ascending=False, na_position='first')

  df_influ_style['style'][i] = df[df == df.iloc[0]].index[0]
```

```python
df_influ_style.head()
```

```python
style_matching_dict = {0: 'americancasual',1: 'formal',2:'dandy',3:'casual',4:'loose',5:'street',6:'street'}
```

```python
for i in range(len(df_influ_style)):
  code_num = df_influ_style['style'].iloc[i]
  df_influ_style['style'].iloc[i] = style_matching_dict[code_num]
  df_influ_style['influ_id'].iloc[i] = df_influ_style['influ_pic_id'].iloc[i][::-1].split('_', 2)[2][::-1]
```

```python
df_influ_style.to_csv('/content/drive/MyDrive/influ_style_tag2.csv')
```

# 남자 Tone

```python
# 모델 로드

model_man_tone = tf.keras.models.load_model('/content/drive/MyDrive/2023 KUIAI 해커톤/man_tone_tag_classificator.h5')
```

```python
df_influ_tone = pd.DataFrame(columns = ['influ_id', 'influ_pic_id','tone'])
df_influ_tone
```

```python
root_file_path = os.listdir('/content/drive/MyDrive/2023 KUIAI 해커톤/influ_man_half')
```

```python
df_influ_tone['influ_pic_id'] = root_file_path
```

```python
for i in range(len(df_influ_tone)):
  dir_path = '/content/drive/MyDrive/2023 KUIAI 해커톤/influ_man_half/'+df_influ_tone['influ_pic_id'][i]
  img = image.load_img(dir_path, target_size=(224,224))
  X = image.img_to_array(img)
  X = np.expand_dims(X,axis=0)
  images = np.vstack([X])
  predictions = np.argmax(model_man_tone.predict(images), axis=1)
  df_influ_tone['tone'][i] = predictions
```

```python
tone_matching_dict = {
    0:'mono',
    1:'neutral',
    2:'pastel',
    3:'toneon'
}
```

```python
for i in range(len(df_influ_tone)):
  code_num = df_influ_tone['tone'].iloc[i][0]
  df_influ_tone['tone'].iloc[i] = tone_matching_dict[code_num]
  df_influ_tone['influ_id'].iloc[i] = df_influ_tone['influ_pic_id'].iloc[i][::-1].split('_', 2)[2][::-1]
```

```python
df_influ_tone.to_csv('/content/drive/MyDrive/man_influ_tone_tag.csv')
```
