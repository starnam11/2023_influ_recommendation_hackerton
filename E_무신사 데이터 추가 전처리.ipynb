```python
# 필요한 모듈 import
import pandas as pd
```

```python
# Database csv 파일 불러오기
data = pd.read_csv('/content/drive/MyDrive/2023 KUIAI 해커톤/Musinsa_database_v2.csv', encoding = 'cp949', index_col = 0)
data
```

```python
# 무신사에서 제공하는 11개 Style Tag > list type으로 변수에 저장
tags_list = ['#아메리칸 캐주얼', '#캐주얼', '#시크', '#댄디', '#포멀', '#걸리시',
             '#골프', '#레트로', '#로맨틱', '#스포츠', '#스트릿']
```

```python
# 기존 DataFrame애 'style' column 추가
data['style'] = 0

# 기존 'tags_list' column에서 11개의 Style Tag에 해당하는 해시태그 추출 > 'style' column에 저장
for i in range(len(data)):
  for style in tags_list:
    if style in data['tags_list'].iloc[i]:
      data['style'].iloc[i] = style
```

```python
data
```

```python
# 성별(gender)로 나누어서 DataFrame csv 파일 각각 저장
# 남자: 1, 여자: 0

df_man = data[data['gender'] == 1]
df_man = df_man.reset_index(drop = True)

df_woman = data[data['gender'] == 0]
df_woman = df_woman.reset_index(drop = True)
```

```python
df_man.to_csv('/content/drive/MyDrive/2023 KUIAI 해커톤/Data/Man/Man.csv', encoding = 'cp949')
df_woman.to_csv('/content/drive/MyDrive/2023 KUIAI 해커톤/Data/Woman/Woman.csv', encoding = 'cp949')
```
