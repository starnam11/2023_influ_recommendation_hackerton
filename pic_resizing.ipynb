# 사진 사이즈 (224 * 224)로의 변환을 위한 코드 

```python
import cv2
from google.colab.patches import cv2_imshow



img = cv2.imread("/content/drive/MyDrive/mans_image_pic/american/26758.jpg")
cv2_imshow(img)
cv2.waitKey()
```

```python
print("img.shape = {0}".format(img.shape))

resize_img = cv2.resize(img, (224, 224))
print("resize_img.shape = {0}".format(resize_img.shape))

cv2_imshow(img)
cv2_imshow(resize_img)
cv2.waitKey()

# 출력
# img.shape = (360, 480, 3)
# resize_img.shape = (300, 300, 3)
```

# 각각의 사진들을 크기 변환 후 각각의 카테고리로 자동 저장

```python
import os
import glob
from PIL import Image


files = glob.glob('/content/drive/MyDrive/mans_image_pic/american/*.jpg')

for f in files:
    img = Image.open(f)
    img_resize = img.resize((224, 224))
    title, ext = os.path.splitext(f)
    img_resize.save( '/content/drive/MyDrive/man_pic/american/' + title[-5:] + '_half' + ext)
```
