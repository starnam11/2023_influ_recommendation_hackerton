```python
# 필요한 모듈 import
import pandas as pd
import os, shutil
```

```python
# 남성, 여성에 해당하는 Database csv 파일 불러오기
man = pd.read_csv('/content/drive/MyDrive/2023 KUIAI 해커톤/Data/Man/Man.csv', encoding = 'cp949', index_col = 0)
woman = pd.read_csv('/content/drive/MyDrive/2023 KUIAI 해커톤/Data/Woman/Woman.csv', encoding = 'cp949', index_col = 0)
```

```python
# '#style'과 이미지 파일을 이동할 폴더명을 dict type으로 매칭하여 변수에 저장
folder_match_dict = {'#캐주얼':'Casual',
 '#댄디':'Dandy',
 '#스트릿':'Street',
 '#골프':'Golf',
 '#아메리칸 캐주얼':'AmericanCasual',
 '#스포츠':'Sports',
 '#로맨틱':'Romantic',
 '#포멀':'Formal',
 '#걸리시':'Girlish',
 '#시크':'Chic',
 '#레트로':'Retro'}
```

```python
# 기존 크롤링한 이미지 파일들이 저장된 폴더 경로
path_before = '/content/drive/MyDrive/2023 KUIAI 해커톤/Data/woman_half/'

# 남성에 해당하는 이미지 파일의 이름들을 list type에 저장
list_jpg = list(woman['image_id'])

# 반복문을 통해 각 이미지 파일이 style에 맞는 폴더로 복사/이동
for i in range(len(list_jpg)):
  num = list_jpg[i]
  line = woman[woman['image_id'] == num]

  style = list(woman[woman['image_id'] == num]['style'])[0]

  if style == '0':
    continue

  # 이동할 이미지 파일 경로
  file_path = path_before + str(num) + '_half.jpg'

  # 이동 시킬 폴더 경로
  move_folder_path = '/content/drive/MyDrive/2023 KUIAI 해커톤/Data/Woman/' + folder_match_dict[style]
  shutil.copy(file_path, move_folder_path)
```
