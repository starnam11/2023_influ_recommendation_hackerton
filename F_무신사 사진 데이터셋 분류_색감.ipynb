```python
# 필요한 모듈 import
import pandas as pd
import os, shutil, ast
```

```python
# 남성, 여성에 해당하는 Database csv 파일 불러오기
man = pd.read_csv('/content/drive/MyDrive/2023 KUIAI 해커톤/Data/Man/Man.csv', encoding = 'cp949', index_col = 0)
woman = pd.read_csv('/content/drive/MyDrive/2023 KUIAI 해커톤/Data/Woman/Woman.csv', encoding = 'cp949', index_col = 0)
```

```python
for i in range(len(man)):
  tag_list = man['tags_list'].iloc[i]
  tag_list = ast.literal_eval(tag_list)
  man['tags_list'].iloc[i] = tag_list
```

```python
# '#style'과 이미지 파일을 이동할 폴더명을 dict type으로 매칭하여 변수에 저장
folder_match_dict = {'#톤온톤':'toneon',
 '#뉴트럴톤':'neutral',
 '#모노톤':'mono',
 '#파스텔':'pastel'
}
```

```python
man['tone'] = 0

for i in range(len(man)):
  if '#톤온톤' in man['tags_list'].iloc[i]:
    man['tone'].iloc[i] = '#톤온톤'
  elif '#뉴트럴톤' in man['tags_list'].iloc[i]:
    man['tone'].iloc[i] = '#뉴트럴톤'
  elif '#모노톤' in man['tags_list'].iloc[i]:
    man['tone'].iloc[i] = '#모노톤'
  elif '#파스텔' in man['tags_list'].iloc[i]:
    man['tone'].iloc[i] = '#파스텔'
```

```python
man['tags_list'].iloc[0]
```

```python
# 기존 크롤링한 이미지 파일들이 저장된 폴더 경로
path_before = '/content/drive/MyDrive/2023 KUIAI 해커톤/Data/man_half/'

# 남성에 해당하는 이미지 파일의 이름들을 list type에 저장
list_jpg = list(man['image_id'])

# 반복문을 통해 각 이미지 파일이 style에 맞는 폴더로 복사/이동
for i in range(len(list_jpg)):
  num = list_jpg[i]
  line = man[man['image_id'] == num]

  tone = list(man[man['image_id'] == num]['tone'])[0]

  if tone == 0:
    continue

  # 이동할 이미지 파일 경로
  img_name = str(num) + '_half.jpg'
  if img_name not in os.listdir(path_before):
    continue 
  file_path = path_before + img_name

  # 이동 시킬 폴더 경로
  move_folder_path = '/content/drive/MyDrive/2023 KUIAI 해커톤/Data/Tone/man/' + folder_match_dict[tone]
  shutil.copy(file_path, move_folder_path)
```
