```python
# 사용할 모듈 import

import os, re
import pandas as pd
import numpy as np
```

```python
# 전처리 진행하는 함수 구현
def cleansing(text):

  # '\n' 및 특수문자 전처리
  text = re.sub(pattern = '\n', repl = ' ', string = text)
  text = re.sub(pattern = '[•;0-9ㄱ-ㅎㅏ-ㅢ+\()/^!?%\xa0=~:\.{2,}]+', repl = '', string = text)
  text = re.sub(pattern = '([_♡❤✌❣♥☀ᆢ⭐�⠀✊❤️❄⛄…✏✨✔⤵️☺️;”“]+)', repl = '', string = text) # 이모티콘 

  # 이모티콘 전처리
  emoji_pattern = re.compile("["
        u"\U00010000-\U0010FFFF"  #BMP characters 이외
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           "]+", flags=re.UNICODE)
  text = re.sub(pattern = emoji_pattern, repl = '', string = text)

  # 이중 스페이스 전처리
  text = re.sub(pattern = re.compile(r'\s+'), repl = ' ', string = text)
  '''
    # Remove emoticons
    def remove_emoji(inputString):
        return inputString.encode('ascii', 'ignore').decode('ascii')
  '''
  return text
```

```python
# 각 인플루언서 별 data.json 파일 접근

'''
1. '인플루언서정보_샘플' 폴더에 있는 100명의 폴더 이름 추출 -> 리스트 형태로 저장
2. 개별 인플루언서 폴더에서 'data_json.csv' 파일 불러오기
3. 
'''

path = '/content/drive/MyDrive/2023 KUIAI 해커톤/Data/인플루언서정보_샘플/'

# 인플루언서 인스타그램 id로 된 폴더 이름 추출 -> 리스트 형태로 저장
inf_id = os.listdir(path)

# 빈 DataFrame 생성
df = pd.DataFrame(columns = ['insta_id','all_text','hashtag','mentioned_id','remained_text'])

# 인플루언서 개별 폴더 경로
for i in range(len(inf_id)):
  path_inf = path + inf_id[i]

  # 인플루언서 별 'json_data.csv' 파일 불러오기
  data = pd.read_csv(path_inf + '/json_data.csv', index_col = 0)

  # 모든 피드 내용 합치기
  all_text = ''

  for j in range(len(data)):
    j_text = data['text'].loc[j]
    all_text = all_text + ' ' + str(j_text)

  # 전처리 함수 적용
  all_text = cleansing(all_text)

  # split
  all_text = all_text.split(' ')

  # 텍스트 분류
  text, hashtag, id = [], [], []

  for word in all_text:
    if '#' in word:
      hashtag.append(word)
    elif '@' in word:
      id.append(word)
    else:
      text.append(word)

  # DataFrame에 추가
  df = df.append({'insta_id':inf_id[i], 'all_text':all_text, 'hashtag':hashtag, 'mentioned_id':id, 'remained_text':text}, ignore_index = True)

# df -> csv 파일로 내보내기
df.to_csv(path + '100명_통합.csv', encoding = 'euc-kr')
```
